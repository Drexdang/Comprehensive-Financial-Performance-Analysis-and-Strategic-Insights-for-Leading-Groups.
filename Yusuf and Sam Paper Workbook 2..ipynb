# Financial Ratios

```python
# Calculate ROE for each bank
df['ROE'] = (df['Profit_For_The_Year'] / df['Shareholders_Equity']) * 100

# Calculate Current Ratio and Debt-to-Equity Ratio
df['Current_Ratio'] = df['Current_Assets'] / df['Current_Liabilities']
df['Debt_to_Equity_Ratio'] = (df['Current_Liabilities'] + df['Long_Term_Liabilities']) / df['Shareholders_Equity']
```

# Profitability Analysis

```python
# Calculate Net Profit Margin: Net Profit / Net Revenue
df['Net_Profit_Margin'] = (df['Profit_For_The_Year'] / df['Net_Revenue']) * 100

# Calculate Operating Expense Ratio: Total Expenses / Net Revenue
df['Operating_Expense_Ratio'] = (df['Total_Expenses'] / df['Net_Revenue']) * 100
```

# Liquidity Analysis

```python
# Calculate Quick Ratio: (Current Assets - Inventory) / Current Liabilities
df['Quick_Ratio'] = df['Current_Assets']   / df['Current_Liabilities']
```

# Assets Utilization Analysis

```python
# Calculate Asset Turnover Ratio: Net Revenue / Average Total Assets
df['Average_Total_Assets'] = (df['Current_Assets'] + df['Non_Current_Assets']) / 2
df['Asset_Turnover_Ratio'] = df['Net_Revenue'] / df['Average_Total_Assets']
```

# Debt Management Analysis

```python
# Calculate Debt Ratio: Total Liabilities / Total Assets
df['Debt_Ratio'] = (df['Current_Liabilities'] + df['Long_Term_Liabilities']) / (df['Current_Assets'] + df['Non_Current_Assets'])

# Calculate Interest Coverage Ratio: Operating Income / Interest Expense
df['Interest_Coverage_Ratio'] = df['Net_Revenue'] / df['Total_Expenses']
```
